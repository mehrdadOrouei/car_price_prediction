```python
import pandas as pd
import numpy as numpy
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
car = pd.read_csv('./quikr_car.csv')
```

```python
car.head()
```

```python
companies=[]
for i in car['company']:
    if i not in companies:
        companies.append(i)
len(companies)
```

```python
c=[]
for i in car['company']:
    c.append(companies.index(i))
c
```

```python
car['company']=c
```

```python
x=[]
for i in car['name']:
    if i not in x:
        x.append(i)
x
```

```python
kms=[]
for i in car['kms_driven']:
    kms.append(i.split("kms"))
kms
```

```python
kms_p=[]
for i in range(len(kms)):
    kms_p.append(int(kms[i][0].replace(",","")))
    # kms_p.append
kms_p
```

```python
car['kms_driven']=kms_p
```

```python
car=car[car['Price']!='Ask For Price']
car["Price"]
```

```python
price=[]
for i in car['Price']:
    price.append(int(i.replace(",","")))
price
```

```python
car['Price']=price
```

```python
car
```

```python
year=[]
for i in car['year']:
    year.append(int(i))
year
```

```python
car['year']=year
```

```python
car=car.drop(['name'],axis=1)
```

```python
x= car.drop(['Price'],axis=1)
```

```python
y=car['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```
